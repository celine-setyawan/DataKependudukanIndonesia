```python
df_merged = pd.merge(df_sensus, df_2016[['Provinsi','Jumlah Penduduk (Ribu)']], how = 'right', 
                     left_on = 'Provinsi', right_on = 'Provinsi')
df_merged.rename(columns={'Jumlah Penduduk (Ribu)': '2016'}, inplace=True)

df_merged = pd.merge(df_merged, df_2017[['Provinsi','Jumlah Penduduk (Ribu)']], how = 'left', 
                     left_on = 'Provinsi', right_on = 'Provinsi')
df_merged.rename(columns={'Jumlah Penduduk (Ribu)': '2017'}, inplace=True)

df_merged = pd.merge(df_merged, df_2018[['Provinsi','Jumlah Penduduk (Ribu)']], how = 'left', 
                     left_on = 'Provinsi', right_on = 'Provinsi')
df_merged.rename(columns={'Jumlah Penduduk (Ribu)': '2018'}, inplace=True)

df_merged = pd.merge(df_merged, df_2019[['Provinsi','Jumlah Penduduk (Ribu)']], how = 'left', 
                     left_on = 'Provinsi', right_on = 'Provinsi')
df_merged.rename(columns={'Jumlah Penduduk (Ribu)': '2019'}, inplace=True)

df_merged.iloc[:, 7:] = df_merged.iloc[:, 7:].astype(float)
# Dikali 1000 sehingga tidak dalam ribu lagi
df_merged.iloc[:, 7:] = np.round((df_merged.iloc[:, 7:] * 1000), 1)

df_merged.to_csv("../jumlah_penduduk_1971-2019.csv", encoding='utf-8', index=False)

df_merged
```
